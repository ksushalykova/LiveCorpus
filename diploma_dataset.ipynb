```python
import pandas as pd
```

```python
data8 = pd.DataFrame(columns=['sentence', 'lang', 'label', 'dataset'])
data8
```

```python
import json
```

```python
f = open('mintaka/data/mintaka_train.json')
data = json.load(f)
for i in range(len(data)):
    lab = data[i]['complexityType']
    if lab == 'comparative':
        l = 1
    else:
        l = 0
    #print(data[i])
    data8.loc[len(data8)] = [data[i]['question'], 'en', l, 'mintaka_train.json']
    #print(data[i]['question'])
    for j in data[i]['translations']:
        data8.loc[len(data8)] = [data[i]['translations'][j], j, l, 'mintaka_train.json']
        #print(j, data[i]['translations'][j])
```
